### Step 1: Data Preprocessing


### Importing the libraries


```python
import pandas as pd
import numpy as np
```

### Importing the dataset


```python
dataset = pd.read_csv(r'C:\Users\prava\100_days_of _ML\datasets\50_Startups.csv')
```

```python
dataset
```

```python
X= dataset.iloc[:,:-1].values
```

```python
X
```

```python
Y= dataset.iloc[:,4].values
```

```python
Y
```

### Encoding Categorical data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder= LabelEncoder()
```

```python
X[:,3]= labelencoder.fit_transform(X[:,3])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
t = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [3])  # Apply OneHotEncoder to column 3
    ],
    remainder='passthrough'  # Keep other columns as they are
)
```

```python
X = t.fit_transform(X)
```

```python
X
```

### Avoiding Dummy Variable Trap

```python
X = X[: , 1:]
```

```python
X
```

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

### Step 2: Fitting Multiple Linear Regression to the Training set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

### Step 3: Predicting the Test set results

```python
y_pred=regressor.predict(X_test)
```

```python
print("Predicted Values:", y_pred)
```

```python
print("Actual Values:", Y_test)
```

## Visualizations:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_dataset = dataset.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_dataset.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, y_pred)
plt.xlabel("Actual Profit")
plt.ylabel("Predicted Profit")
plt.title("Actual Profit vs. Predicted Profit")
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red')  # Add a diagonal line for reference
plt.show()
```

```python
# Residual plot
residuals = Y_test - y_pred
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Profit")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
# Distribution of residuals
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True)
plt.title("Distribution of Residuals")
plt.xlabel("Residuals")
plt.show()
```

## Evaluation

```python
from sklearn.metrics import r2_score
```

```python
R_square = r2_score(Y_test,y_pred)
```

```python
print(f"R-squared: {R_square}")
```

```python
print(f"Type of X after ColumnTransformer: {type(X)}")
```

#### **Results:** This analysis aimed to predict startup profits using a multiple linear regression model. After preprocessing the data, including one-hot encoding categorical features, the model was trained and evaluated. The model demonstrated excellent performance, achieving an R-squared of 0.9347. This suggests that the model can explain approximately 93.47% of the variance in startup profits. The scatter plot of actual vs. predicted values and the residual plots further confirmed the model's accuracy and the validity of the linear regression assumptions.
